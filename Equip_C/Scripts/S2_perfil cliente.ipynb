```python
# Variable num茅rica y variable binaria
# Example of the Student's t-test
from scipy.stats import ttest_ind
data1 = cliente['housing']
data2 = cliente['age']
stat, p = ttest_ind(data1, data2)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
	print('Probably the same distribution')
else:
	print('Probably different distributions')

## por el resultado no hay una influencia de la edad en la contrataci贸n de hip贸teca
```

### ANALISIS MULTIVARIANTE

```python
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm

# Convertir variable binaria a proporciones por grupo
cliente['housing_mean'] = cliente.groupby(['job', 'marital', 'education'])['housing'].transform('mean')

# Modelo ANOVA
modelo = ols('housing_mean ~ C(job) + C(marital) + C(education)', data=cliente).fit()
anova_results = anova_lm(modelo)

print(anova_results)
```

## LOAN

```python
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm

# Convertir variable binaria a proporciones por grupo
cliente['loan_mean'] = cliente.groupby(['job', 'marital', 'education'])['loan'].transform('mean')

# Modelo ANOVA
modelo = ols('loan_mean ~ C(job) + C(marital) + C(education)', data=cliente).fit()
anova_results = anova_lm(modelo)

print(anova_results);
```

CONCLUSIONES:
1) Todas las variables categ贸ricas (job, marital, education) tienen un efecto significativo sobre la variable dependiente (loan), dado que los p-valores son <0.05
2) La variable job tiene el mayor impacto en la variabilidad explicada (=2606.75 y suma de cuadrados m谩s alta: 
33.57).
3) Residualmean_sq = 0.001171. La variabilidad residual es peque帽a, lo que indica que el modelo ANOVA ajusta bien los datos.
