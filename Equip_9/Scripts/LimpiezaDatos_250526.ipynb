### Librerías

```python
import os
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")#ignorar advertencias
```

### Carga de datos

```python
df = pd.read_csv("../Data/df_original_250526.csv")
df
```

### Duplicados exactos

```python
duplicados = df[df.duplicated()]
print(f"Filas duplicadas exactas: {len(duplicados)}")
duplicados
```

Se han detectado 8 filas duplicadas exactas, las eliminamos por que no aportan informacion nueva al analisis y pueden sesgar resultados

```python
# Eliminar duplicados exactos
df = df.drop_duplicates()

# Confirmar que se han eliminado
print(df.shape)
```

### IDs dulpicados

```python
ids_duplicados = df[df['id'].duplicated()]
print(f"Ids duplicados: {len(ids_duplicados)}")

total_ids_duplicados = df['id'][df['id'].duplicated(keep=False)]
duplicados_id = df[df['id'].isin(total_ids_duplicados)]
print(f"Total de filas con ids duplicados: {len(duplicados_id)}")
duplicados_id.sort_values('id')
```

### Valores nulos

```python
df.isnull().sum()
```

Tratamiento de nulos tras acuerdo grupal
- age --> media age
- marital --> moda marital
- education --> unknown
- housing --> moda housing

```python
print(df['default'].mode())         # 0    False --> Solo hay una moda
```

```python
# Rellenar valores nulos
df['default'].fillna(df['default'].mode()[0], inplace=True)
```

```python
# Comprobar
df.isnull().sum()
```

### Convertir columnas booleanas

```python
# Detectar columnas con solo 'yes' y 'no'
for col in df.columns:
    unique_vals = df[col].dropna().unique()
    if set(unique_vals).issubset({'yes', 'no'}):
        print(f"La columna '{col}' contiene solo 'yes' y 'no'")
```

```python
# Convertir columnas binarias 'yes'/'no' a booleanas
bool_columns = ['default', 'housing', 'loan', 'deposit']
for col in bool_columns:
    df[col] = df[col].map({'yes': True, 'no': False})
print(df.dtypes)
```

### Corrección columna `age`

```python
# Convertir 'age' a entero 
df['age'] = pd.to_numeric(df['age'], errors='coerce').fillna(0).astype(int)
print(df.dtypes)
```

### Corrección columna `month`

```python
# Asegurar que 'month' tenga solo la primera letra en mayúscula
df['month'] = df['month'].str.strip().str.lower().str.capitalize()
df
```

### Crear columna `date`

```python
# Crear columna 'date' combinando 'day' y 'month' (asumiendo año 2024)
df['date'] = pd.to_datetime(
    df['day'].astype(str) + '-' + df['month'] + '-2024',
    dayfirst=True
)
df
```

### Guardar el DataFrame limpio

```python
df.to_csv("../Data/df_250526.csv")
```
