## Modificación datos

### Eliminar duplicados exactos

```python
# Eliminar duplicados exactos
df = df.drop_duplicates()

# Confirmar que se han eliminado
print(df.shape)
```

### Rellenar valores nulos
Con SimpleImputer rellenamos los valores categóricos con la moda y el valor numérico con la mediana.

```python
# Compruebo modas y medianas antes de imputar datos
for col in ['marital', 'education', 'default', 'housing']:
    print(f"Moda de '{col}': {df[col].mode()[0]}")
    print(df[col].value_counts())
    print()

mediana_age = df['age'].median()
print(f"Mediana de 'age': {mediana_age}")
print((df['age'] == mediana_age).sum())
```

```python
cat_imputer = SimpleImputer(strategy="most_frequent")
df[['marital', 'education', 'default', 'housing']] = cat_imputer.fit_transform(df[['marital', 'education', 'default', 'housing']])

num_imputer = SimpleImputer(strategy="median")
df[['age']] = num_imputer.fit_transform(df[['age']])

# Comprobar nulos
print(df.isnull().sum())
print("Total nulos = ", df.isnull().sum().sum())
```

```python
# Compruebo modas y medianas después de imputar datos
for col in ['marital', 'education', 'default', 'housing']:
    print(f"Moda de '{col}': {df[col].mode()[0]}")
    print(df[col].value_counts())
    print()

mediana_age = df['age'].median()
print(f"Mediana de 'age': {mediana_age}")
print((df['age'] == mediana_age).sum())
```

### Reasignación de nuevas IDs a filas con IDs duplicadas

```python
df.shape
```

```python
# COPIA DE SEGURIDAD
df_original = df.copy()

# 1. Detectar duplicados por ID (no idénticos)
total_ids_duplicados = df[df['id'].duplicated(keep=False)]
duplicados_no_identicos = total_ids_duplicados.groupby('id').filter(lambda x: not x.duplicated().all())

# 2. Reasignar nuevos IDs a una copia segura
duplicados_no_identicos = duplicados_no_identicos.copy()
max_id_actual = df['id'].max()
duplicados_no_identicos['id'] = range(max_id_actual + 1, max_id_actual + 1 + len(duplicados_no_identicos))

# 3. Eliminar registros originales con IDs problemáticos
ids_a_eliminar = duplicados_no_identicos['id'].duplicated(keep=False)  # Ya reasignados → no están en el original
ids_originales = total_ids_duplicados['id'].unique()
df = df[~df['id'].isin(ids_originales)].reset_index(drop=True)

# 4. Añadir filas corregidas
df = pd.concat([df, duplicados_no_identicos], ignore_index=True)

# 5. Verificación final
print(f"Filas finales en el dataset corregido: {df.shape[0]}")
print(f"Total de IDs únicos: {df['id'].nunique()}")
```

```python
# Verificación de duplicados
total_ids_duplicados = df[df['id'].duplicated(keep=False)]
print(f"IDs duplicados post corrección: {total_ids_duplicados['id'].nunique()}")
print(f"Filas con IDs duplicados post corrección: {len(total_ids_duplicados)}")
```

### Convertir columnas booleanas

```python
# Detectar columnas con solo 'yes' y 'no'
for col in df.columns:
    unique_vals = df[col].dropna().unique()
    if set(unique_vals).issubset({'yes', 'no'}):
        print(f"La columna '{col}' contiene solo 'yes' y 'no'")
```

```python
# Convertir columnas binarias 'yes'/'no' a booleanas
bool_columns = ['default', 'housing', 'loan', 'deposit']
for col in bool_columns:
    df[col] = df[col].map({'yes': True, 'no': False})
print(df.dtypes)
```

### Corrección columna `age`

```python
# Convertir 'age' a entero 
df['age'] = pd.to_numeric(df['age'], errors='coerce').fillna(0).astype(int)
print(df.dtypes)
```

### Corrección columna `month`

```python
# Asegurar que 'month' tenga solo la primera letra en mayúscula
df['month'] = df['month'].str.strip().str.lower().str.capitalize()
df
```

### Crear columna `date`

```python
# Crear columna 'date' combinando 'day' y 'month' (asumiendo año 2024)
df['date'] = pd.to_datetime(
    df['day'].astype(str) + '-' + df['month'] + '-2024',
    dayfirst=True
)
df
```
